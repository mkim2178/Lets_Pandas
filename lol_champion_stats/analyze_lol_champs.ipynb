# LOL champion stats!

```python
import pandas as pd
import numpy as np

# every champion's stat is based on level 1

champ_stats = pd.read_csv("ver_15.7.1_champion_stats.csv")
```

### Find champions that satisfies a following condition: *hp less than 600*

```python
hp_less_than_600 = champ_stats[champ_stats["hp"] < 600]
print("number of champions that hp < 600:", hp_less_than_600.shape[0])
hp_less_than_600
```

### Find champions that satisfies a following condition: *hp less than 600 AND move speed greater than 330*

```python
hp_less_than_600_and_move_speed_greater_than_330 = champ_stats[(champ_stats["hp"] < 600) & (champ_stats["move_speed"] > 330)]
print("number of champions that hp < 600 and move speed > 330:", hp_less_than_600_and_move_speed_greater_than_330.shape[0])
hp_less_than_600_and_move_speed_greater_than_330
```

### Find champions that name starts with "H"

```python
champions_starts_with_H = champ_stats[champ_stats["name"].str.startswith("H")]
champions_starts_with_H
```

### Create a new column calld "hp_plus_mp" which is the sum of each champion's hp and mp

```python
hp_plus_mp = champ_stats["hp"] + champ_stats["mp"]
champ_stats["hp_plus_mp"] = hp_plus_mp
champ_stats.head()
```

### Rename a column 'hp_plus_mp' to 'hp+mp'

```python
champ_stats = champ_stats.rename(columns={"hp_plus_mp" : "hp+mp"})
champ_stats.head()
```

### Remove the column 'hp+mp' by using .drop

```python
champ_stats = champ_stats.drop("hp+mp", axis="columns")
champ_stats.head()
```

### Get mean and max of hp by using numpy functions

```python
get_all_HP = champ_stats["hp"]
HP_mean = np.mean(get_all_HP)
HP_max = np.max(champ_stats["hp"])
HP_max
```

### Use .describe() to get statistics of every column

```python
champ_stats.describe()
```

### Get 10 random champions name by using .sample() function

```python
ten_random_champs = []
for i in range(10):
    champ = champ_stats.sample()["name"].values[0]
    ten_random_champs.append(champ)
ten_random_champs
```

### Get 3 random champions with name, title, and hp. Sample **with replacement** and champions should satisfies  following condition: hp greater than 600

```python
champ_stats[champ_stats["hp"] > 600].sample(3, replace = True).iloc[:, 0:3]
```
